```
file_location = "/FileStore/tables/winequality_modified.csv"
file_type = "csv"
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
df = spark.read.format(file_type) \
 .option("inferSchema", infer_schema) \
 .option("header", first_row_is_header) \
 .option("sep", delimiter) \
 .load(file_location)
display(df)
```

```
df = df.drop('quality')
df.printSchema()
```

```
 df=df.withColumnRenamed('Wine quality','label')
```

```
df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```
df_clean = df.na.drop()
df_clean.describe().show()
```

```
df_clean.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```
from pyspark.ml.feature import VectorAssembler, VectorIndexer, StringIndexer, OneHotEncoder
```

```
type_indexer = StringIndexer(inputCol='type', outputCol='typeIndex')
df_ = type_indexer.fit(df_clean).transform(df_clean)
```

```
df_.show(10)
```

```
encoder = OneHotEncoder(inputCols=['typeIndex'],outputCols=['type_vec'])
df_ = encoder.fit(df_).transform(df_)
df_.show(10)
```

```
df_.columns
```

```
assembler = VectorAssembler(inputCols=['fixed acidity',
 'volatile acidity',
 'citric acid',
 'residual sugar',
 'chlorides',
 'free sulfur dioxide',
 'total sulfur dioxide',
 'density',
 'pH',
 'sulphates',
 'alcohol',
 'type_vec'], outputCol='features')
df_ = assembler.transform(df_)
```

```
df_.show(10)
```

```
df_.select(['features','label']).show(10,False)
```

```
model_df=df_.select(['features','label'])
model_df.show()
```

```
 train_df,test_df=model_df.randomSplit([0.75,0.25])
```

```
print(train_df.count())
```

```
print(test_df.count())
```

```
train_df.groupBy('label').count().show()
```

```
test_df.groupBy('label').count().show()
```

```
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
model=RandomForestClassifier(labelCol='label')
paramGrid = ParamGridBuilder().addGrid(model.numTrees,[5,10]).addGrid(model.maxDepth,[2,3,5]).build()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator=BinaryClassificationEvaluator(rawPredictionCol= 'rawPrediction',labelCol='label')
```

```
cv = CrossValidator(estimator=model, estimatorParamMaps=paramGrid, evaluator=evaluator, numFolds=33)
```

```
rf_model = cv.fit(train_df)
```

```
test_results=rf_model.transform(test_df)
test_results.show()
```

```
test_results.groupBy('prediction').count().show()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
rf_auc=BinaryClassificationEvaluator(labelCol='label').evaluate(test_results)
print(rf_auc)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
rf_acc= MulticlassClassificationEvaluator(labelCol='label', metricName ='accuracy').evaluate(test_results)
print(rf_acc)
```

```
bestModel = rf_model.bestModel
```
